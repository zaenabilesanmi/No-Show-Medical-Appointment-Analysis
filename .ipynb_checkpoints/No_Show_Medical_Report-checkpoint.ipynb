Project: Medical Appointment No-shows

Exploratory Analysis of patients attitude to medical appointments in Brazilian hospitals

Introduction "No-shows" is a term used to describe whether or not patients attend their medical appointment in Brazilian Hospitals. This project seeks to investigate factors that could affect patients showing up or not for appointment in these hospitals.

The dataset for this analysis contains 110,527 appointments (rows) in total as 14 defining characteristics in specific context. While seeking to understand the impact of each variables on patients showing up for their appointment, "No-show" and "Scholarship" appears to be one of the most important of all variables. The other variables that were investigated in the dataset include:

PatientId - specific to each patient

AppointmentID - specific to each patient

Gender - male or female

Appointment Day - the day they need to visit the doctor

Schedule Day -the day the patient got booked for the appointment

Age - how old the patient is

Neighbourhood - area of the appointment

Scholarship -whether or not the patient is enrolled in Brasilian free healthcare scheme

Hipertension - whether or not the patient have hypertension

Diabetes - to show whether or not the patient suffers diabetes

Alcoholis - to show whether or not the patient is alcoholic

Handcap - to show whether or not the person is physically challenged

SMS_received - to show whether or not the patient received the reminder for the appointment

No-show (Present or Absent)

Questions

- What is the effect of sending SMS and patient showing up on the day of appointment? 
- Did scholarship affect patients showing up for appointment? 
- What age range showed up more?
